## Case study  - Football Data Analysis

We have the football data of a popular premier league which asseses the differnet qualities in a football player. We need to grill this data and report the important findings from the grilling and drilling exercise. We are also supposed to derive useful insights about the whole dataset.


```python
# Import the libraries 
import numpy as np
import pandas as pd
import seaborn as sns  
import matplotlib.pyplot as plt  # seaborn is based on matplotlib
sns.set(color_codes=True) # adds a nice background to the graphs
%matplotlib inline 
# tells python to actually display the graphs
```

### <a id = "link1"></a>Overview of the data

```python
# Read the Data 
df = pd.read_excel('Football Players.xlsx')
```

```python
# View first 3 rows of data 
df.head(3) 
```

```python
# understand shape and size of data from 
print(df.shape)
print(df.size)
```

```python
df.info()
print()
```

```python
# Conditions within dataframe 
pd=df[df['Sprint speed'] > 70]
pd.sortby(ascending= False)
```

```python
#Unique football positions
df['Preferred Positions'].unique()
```

```python
#Total ST
df[df['Preferred Positions']=='ST']
```

```python
pd=df.groupby('Nationality').size().sort_values(ascending = False)
sns.set({'figure.figsize': (8,6)})
sns.histplot(df['Nationality'], kde=True);
```

```python
#Total players in their respective age gp
age_under20=df[df['Age']<=25]
age_25to30=df[(df['Age']>25)&(df['Age']<30)]
age_above30=df[df['Age']>=30]
var=np.array([age_under20['Name'].count(),age_25to30['Name'].count(),age_above30['Name'].count()])
label=["<=25",">25 & <30",">=30"]
plt.title("Total players in their respective age gp")
plt.pie(x=var,labels=label,autopct="%.2f%%")
plt.show()
```

```python
#TOTAL NO OF NATIONS
df['Nationality'].nunique()
```

```python
df['Preferred Positions'].value_counts().head(10)
```

```python
sns.jointplot(df['Standing tackle'],df['Sliding tackle'], kind="hex");
```

```python
#Most Valuable club
k1=df.groupby('Club')['Value ($)'].sum().sort_values(ascending=False)
```

**Real Madrid CF is the most valuable team in the world with a squad of worth 7.18 billion U.S.D**

```python
#Highest paying club
df.groupby('Club')['Wage ($)'].sum().sort_values(ascending=False)
```

**FC Barcelona are the Premier leagues's highest paying club, spending around 4 million U.S.D**
